```python
#1-
from datetime import datetime

# 1- Creo una clase llamada empleados, con sus respectivos atributos.
# y despues le asigno metodos que va a recibir 
#y los imprimo en el metodo descripcion

class Empleados:
    def __init__(self,nombre,apellido,edad,posicion,año,vacaciones,herramientas, salario):
        self.nombre = nombre
        self.apellido = apellido
        self.edad = edad
        self.posicion = posicion #posicion en la empresa
        self.año = año #año de ingreso
        self.vacaciones = vacaciones #cantidad de dias de vacaciones que le quedan
        self.herramientas = herramientas #herramientas que el empleado sabe utilizar
        self.salario = salario #agrege el salario del punto 6

    def descripcion(self):
        return(f"Empleada: {self.nombre} {self.apellido}\n"
                f"Posición: {self.posicion}\n"
                f"Año de ingreso: {self.año}\n"
                f"Días de vacaciones restantes: {self.vacaciones}\n"
                f"Salario: ${self.salario}") #agregue el salario en el punto 6

    

# 2- Verifico las herramientas que el empleado tiene en su lista y devuelvo un mensaje específico basado en las herramientas.

    def cambiando_posicion(self):
        mensajes = []
        for herramienta in self.herramientas:
            if herramienta == "python": #verifica si es igual a "Python" 
                    mensajes.append(f"Excelente {self.nombre} ! Es buenisimo que sepas Python")
            elif herramienta == "excel": #verifica si es igual a "Excel"
                    mensajes.append(f"Excelente {self.nombre} que sepas Excel, pero te sugerimos que aprendas Python.")
        return "\n".join(mensajes) #uno los elementos de la lista mensajes con un salto de línea entre cada elemento, para que cada mensaje aparezca en un renglon nuevo.
        
 # 3- Calculo la antigüedad de la empleada en años y meses desde el año de ingreso hasta la fecha actual:

    def calcular_antiguedad(self):
        
        fecha_actual = datetime.now() #pido la fecha actual
        año_actual = fecha_actual.year #separo el año
        mes_actual = fecha_actual.month #separo el mes

        años_antiguedad = año_actual - self.año #indico que los años de antiguedad = al año actual menos el año de ingreso que estaba antes indicado
        meses_antiguedad = mes_actual #

        if meses_antiguedad < 0:
            años_antiguedad -= 1
            meses_antiguedad += 12

        return f"{self.nombre} lleva {años_antiguedad} años y {meses_antiguedad} meses en la empresa."

# 4- Método para aumentar los dias de vacaciones que tiene la empleada

    def solicitar_aumento_vacaciones(self, dias_solicitados):
        if dias_solicitados > 0: #verifico que los dias pedidos sean mayor a cero
            self.vacaciones= self.vacaciones + dias_solicitados # Actualizo los días de vacaciones pedidos
            return (f"Aprobado. Pediste aumentar tus dias de vacaciones.Ahora tienes {dias_solicitados} dias más de vacaciones. Ahora en total tienes {self.vacaciones} días de vacaciones.")
        else:
            return ("Error: La solicitud debe ser un número positivo.")
        
# 5- Metodo para registrar nuevas herramientas de la empleada en la lista de habilidades

    def registrar_herramienta(self, nueva_herramienta):
        if nueva_herramienta.lower() not in (herramienta.lower() for herramienta in self.herramientas):  # Valido si la nueva herramienta está o no en la lista
            self.herramientas.append(nueva_herramienta) # Agregar la nueva herramienta a la lista
            return f"Herramienta '{nueva_herramienta}' quedó registrada exitosamente para {self.nombre}."
        else:
            return f"La herramienta '{nueva_herramienta}' ya está registrada para {self.nombre}." #mensaje por si la herramienta ya estaba en la lista

# 6- Cambiar la posicion de la empleada y su salario
# ojo ! debo agregar el atributo salario, a la clase Empleados

    def cambiar_posicion(self, nueva_posicion, nuevo_salario):
            self.posicion = nueva_posicion
            self.salario = nuevo_salario
            return (f"¡Felicidades, {self.nombre}!\n" 
                    f"Has sido promovida a la posición de '{self.posicion}' con un salario de ${self.salario:.2f}.\n"
                    f"¡Sigue con el excelente trabajo!")
    
# 7- Tomar vacaciones
    def tomar_vacaciones(self):
        while True:
            try:
                dias_solicitados = int(input(f"{self.nombre}, ¿cuántos días de vacaciones deseas tomar? "))
                if dias_solicitados <= 0:
                    print("Error: Debes solicitar al menos un día de vacaciones.")
                elif dias_solicitados > self.vacaciones:
                    print(f"Error: No tienes suficientes días de vacaciones. Te quedan {self.vacaciones} días disponibles.")
                else:
                    self.vacaciones= self.vacaciones - dias_solicitados
                    return (f"Aprovado. Pediste {dias_solicitados} dias de vacaciones. Ahora te quedan {self.vacaciones} días de vacaciones.")
            except ValueError:
                print("Error: Por favor, ingresa un número válido.")

# 8- Solicitar aumento de salario
    def solicitar_aumento_salario(self):
        while True:
            try:
                porcentaje_aumento = float(input(f"{self.nombre}, ¿qué porcentaje de aumento deseas solicitar? "))
                if porcentaje_aumento < 0:
                    print("Error: El porcentaje de aumento no puede ser negativo.")
                else:
                    nuevo_salario = self.salario * (1 + porcentaje_aumento / 100)
                    self.salario = nuevo_salario
                    return (f"Solicitud exitosa. Pediste un aumento de {porcentaje_aumento} %. Tú nuevo salario es ${self.salario:.2f}.")
            except ValueError:
                print("Error: Por favor, ingresa un número válido.")

# 9- Realizar nueva evaluacion 
    def realizar_evaluacion(self):
        opciones_validas = ["excelente", "buena", "regular", "mala"] #defino las opciones en una lista
        while True:
            calificacion = input(f"{self.nombre}, ¿cómo calificas tu desempeño? Opciones: excelente, buena, regular, mala: ").lower()
            if calificacion in opciones_validas:
                if calificacion == "excelente":
                    return "¡Increíble desempeño! Sigue con tú excelente trabajo."
                elif calificacion == "buena":
                    return "Buen trabajo. Sigue esforzándote para llegar a la excelencia."
                elif calificacion == "regular":
                    return "Hay áreas para mejorar. Trabaja en tus puntos débiles."
                elif calificacion == "mala":
                    return "Es importante reflexionar y mejorar. ¡Tú puedes hacerlo!"
            else:
                print("Error: Por favor, ingresa una calificación válida. Recuerda tus opciones son: excelente, buena, regular, mala).")


empleada1= Empleados("Maria Eleonora", "Soneira",41, "Analista de datos", 2020, 15,["python", "excel"], 1500)

#1- Imprimo la descripcion de la empleada 1
print(empleada1.descripcion())
print("********************")
#2- Imprimo los mensajes basado en las herramientas.
print(empleada1.cambiando_posicion())
print("********************")
#3- Calculo la antiguedad de la empleada
print(empleada1.calcular_antiguedad())
print("********************")
#4- Aumentar dias de vacaciones:
print(empleada1.solicitar_aumento_vacaciones(5))  # Solicitud válida
print("********************")
#5-agregar nueva herramienta (habilidad)
print(empleada1.registrar_herramienta("Power BI")) # Nueva herramienta
print(empleada1.registrar_herramienta("Tableau"))  # Nueva herramienta
print(empleada1.registrar_herramienta("python"))  # Herramienta ya registrada
print(f"Herramientas actuales de {empleada1.nombre}: {empleada1.herramientas}") #la lista actualizada
#6-Cambio posicion y salario
print("********************")
print(empleada1.cambiar_posicion("Data Engineer",3000))
print("********************")
#7- la empleada toma dias de vacaciones, los pide con input, por eso aca no le pongo el valor
print(empleada1.tomar_vacaciones())
print("********************")
# 8- Aumento de salario, la empleada lo pide con input, por eso aca no le pongo el valor
print(empleada1.solicitar_aumento_salario())
print("********************")
# 9- Realiza evaluación. La empleada nos dice como fue su desempeño con un input, por eso aca no le pongo el valor
print(empleada1.realizar_evaluacion())
print("********************")
```

```python
class Gerente(Empleados):
    def __init__(self, nombre, apellido, edad, posicion, año, vacaciones, herramientas, salario, departamento, empleados_a_cargo):
        super().__init__(nombre, apellido, edad, posicion, año, vacaciones, herramientas, salario)
        self.departamento = departamento
        self.empleados_a_cargo = empleados_a_cargo  # Lista de empleados a cargo.

    def asignar_tareas(self, tarea, empleado):
        return f"La gerente {self.nombre} ha asignado la tarea '{tarea}' a la empleada {empleado}."

    def revisar_desempeno_empleado(self, empleado):
        return f"La gerente {self.nombre} está revisando el desempeño de la empleada {empleado}."

    def gestionar_empleados(self, accion, empleado):
        if accion.lower() == "contratar":
            if empleado not in self.empleados_a_cargo:
                self.empleados_a_cargo.append(empleado)
                return f"La empleada {empleado} ha sido contratada por el gerente {self.nombre}."
            else:
                return f"La empleada {empleado} ya está en la lista de empleados a cargo."
        elif accion.lower() == "despedir":
            if empleado in self.empleados_a_cargo:
                self.empleados_a_cargo.remove(empleado)
                return f"La empleada {empleado} ha sido despedida por el gerente {self.nombre}."
            else:
                return f"La empleada {empleado} no está en la lista de empleados a cargo."
        else:
            return "Acción no válida. Por favor, usa 'contratar' o 'despedir'."


gerente1 = Gerente(
    "Maria Eleonora", "Soneira", 41, "Gerente de Proyectos", 2015, 20, ["Gestión", "Planificación"], 70000,
    "Tecnología", ["Yanelis Serrano Taylor", "Rocio Martinez", "Pepita Sanchez"]
)
print("********************")
# Descripción del gerente
print(gerente1.descripcion())
print("********************")
# Asignar una tarea a un empleado
print(gerente1.asignar_tareas("Preparar analisis de ventas", "Rocio Martinez"))
print("********************")
# Revisar el desempeño de un empleado
print(gerente1.revisar_desempeno_empleado("Yanelis Serrano Taylor"))
print("********************")
# Contratar un nuevo empleado
print(gerente1.gestionar_empleados("contratar", "Alba Toro"))
print("********************")
# Intentar contratar a un empleado ya en la lista
print(gerente1.gestionar_empleados("contratar", "Rocio Martinez"))
print("********************")
# Despedir a un empleado
print(gerente1.gestionar_empleados("despedir", "Pepita Sanchez"))
print("********************")
# Intentar despedir a un empleado que no está en la lista
print(gerente1.gestionar_empleados("despedir", "Pepa Garcia"))
print("********************")
# Acción no válida
print(gerente1.gestionar_empleados("promover", "Rocio Martinez"))
```
