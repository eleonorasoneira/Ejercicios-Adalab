```python

class Ecommerce:
    def __init__(self, nombre):
        self.nombre = nombre
        self.productos = {}
        self.clientes = {}
        self.inventario={}
        self.ventas =[]
        self.clientes_ventas = []
        self.ganancias_totales = 0
        self.compras_por_cliente = {}
    
        
    def agregar_producto(self, id_producto, nombre, precio, stock):
        if id_producto not in self.productos:
            self.productos[id_producto] = [nombre,precio,stock]
            print(f"Nuevo Producto añadido correctamente. ID Producto: {id_producto}, producto: {nombre}, Precio: ${precio}, Stock: {stock}" )
        else:
            print(f"El producto {id_producto} ya esta en el diciconario")

    def eliminar_producto(self,id_producto):
        if id_producto in self.productos:
            self.productos.pop(id_producto)
            print (f"Producto {id_producto} eliminado correctamente.")
        else:
            print (f"Producto {id_producto} no esta en el diccionario")

    def agregar_cliente(self, id_cliente, nombre, email):
        if id_cliente not in self.clientes:
            self.clientes[id_cliente] = [nombre,email]
            print(f"Nuevo cliente añadido. Cliente Número {id_cliente}. El nombre de tu nuevo cliente es: {nombre} y su correo electronico es {email}")
        else:
            print(f"El cliente {id_cliente} ya esta en el diciconario")

    def realizar_venta(self,id_cliente, id_producto, cantidad):
        if id_producto in self.productos and id_cliente in self.clientes:
            if self.productos[id_producto][2] >= cantidad:
                id_venta = len(self.ventas) +1
                self.ventas.append((id_venta,id_cliente, id_producto, cantidad))
                self.productos[id_producto][2] -= cantidad
                print (f"Enhorabuena! Venta número: {id_venta} Realizada con exito! \nDetalles de la venta: Has vendido {cantidad} unidades, del producto ID: {id_producto} al cliente ID: {id_cliente}")
            else:
                print(f"No hay suficiente cantidad de ese producto para poder realizar la venta")
        else:
            print("El producto o cliente que buscas no existe")

    def obtener_inventario(self):
        for id_producto, producto_info in self.productos.items():
            nombre,precio,stock = producto_info
            sum_vendido = sum(
            venta[3] for venta in self.ventas if venta[2] == id_producto
            )# Calcula el total vendido para este producto
            self.inventario[id_producto]={
                "Nombre": nombre,
                "Precio": precio,
                "Stock Actual": stock,
                "Stock vendido" :sum_vendido,
                "Stock total": stock + sum_vendido,
                }
        return self.inventario

    def obtener_clientes(self, con_compras=False):
        for id_cliente, cliente_info in self.clientes.items():
            nombre, email = cliente_info
            if con_compras:
                # Calcular el total de compras del cliente
                total_compras = sum(
                    venta[3] for venta in self.ventas if venta[1] == id_cliente
                )
                # Agregar el total de compras al diccionario de clientes
                self.clientes[id_cliente] = {
                    "Nombre": nombre,
                    "Email": email,
                    "Total Compras": total_compras
                }
            else:
                # Solo agregar nombre y email si no se requiere el total de compras
                self.clientes[id_cliente] = {
                    "Nombre": nombre,
                    "Email": email
                }
        
        return self.clientes
    
    def actualizar_producto(self, id_producto, nombre=None, precio=None, stock=None):
        if id_producto in self.productos:
            # Actualizar los valores solo si se pasan nuevos valores
            if nombre is not None:
                self.productos[id_producto][0] = nombre
            if precio is not None:
                self.productos[id_producto][1] = precio
            if stock is not None:
                self.productos[id_producto][2] = stock
            print(f"Producto {id_producto} actualizado con éxito.")
            return self.productos[id_producto]
        else:
            print(f"El producto con ID {id_producto} no existe.")
            return None
    
    def buscar_producto_por_nombre(self, nombre):
        return [
            (id_producto, producto_info)
            for id_producto, producto_info in self.productos.items()
            if nombre.lower() in producto_info[0].lower()
        ]
    def calcular_ganancias(self):
        for venta in self.ventas:
            id_producto = venta[2]
            cantidad = venta[3]
            precio = self.productos[id_producto][1]  # Obtener el precio del producto
            self.ganancias_totales += precio * cantidad
        return self.ganancias_totales
    
    def obtener_clientes_con_mas_compras(self, n=5): ##por defecto que devuelva los mejores 5 clientes que hicieron mas ventas
        for venta in self.ventas:
            id_cliente = venta[1]
            cantidad = venta[3]
            if id_cliente in self.compras_por_cliente:
                self.compras_por_cliente[id_cliente] += cantidad
            else:
                self.compras_por_cliente[id_cliente] = cantidad
        
        # Ordenar a los clientes por el total de compras (de mayor a menor)
        
        mejores_clientes = sorted(self.compras_por_cliente.items(), key=lambda x: x[1], reverse=True)[:n]
        return mejores_clientes
        
    
#Le pongo nombre a mi ecommerce
mi_tienda = Ecommerce("Mercadona")
# Imprimir el nombre de la tienda
print(mi_tienda.nombre)
print("*****************************")

#Agrego productos a mi diccionario de productos
mi_tienda.agregar_producto(1, "Patata", 2, 50)
mi_tienda.agregar_producto(2,"Pan", 2, 50)
mi_tienda.agregar_producto(3,"Queso", 2, 50)
mi_tienda.agregar_producto(4, "Tomate", 2, 50)
mi_tienda.agregar_producto(5,"Fideos", 2, 50)
mi_tienda.agregar_producto(6,"Agua", 2, 50)
mi_tienda.agregar_producto(7,"Pizza", 2, 50)
print("*****************************")

# Verifico el contenido del diccionario
print("Tu tienda actualizada es:",mi_tienda.productos)
print("*****************************")

#Elimino un producto de la tienda
mi_tienda.eliminar_producto(2)
print("*****************************")

#Agrego un cliente a mi tienda
mi_tienda.agregar_cliente(1, "Maria", "maria@gmail.com")
mi_tienda.agregar_cliente(2, "Eleonora", "eleonora@gmail.com")
mi_tienda.agregar_cliente(3, "Florencia", "florencia@gmail.com")
mi_tienda.agregar_cliente(4, "Leticia", "leticia@gmail.com")
mi_tienda.agregar_cliente(5, "Sebastian", "sebastian@gmail.com")
mi_tienda.agregar_cliente(6, "Esteban", "esteban@gmail.com")
mi_tienda.agregar_cliente(7, "Jim", "jim@gmail.com")
print("*****************************")

#Realizo una venta en mi tienda, al cliente 1, del producto 3, en cantidad:5
mi_tienda.realizar_venta(1,1,10)
print("----------------------------")
mi_tienda.realizar_venta(2,3,11)
print("----------------------------")
mi_tienda.realizar_venta(3,4,12)
print("----------------------------")
mi_tienda.realizar_venta(4,4,13)
print("----------------------------")
mi_tienda.realizar_venta(5,6,14)
print("----------------------------")
mi_tienda.realizar_venta(6,7,15)
print("----------------------------")
mi_tienda.realizar_venta(7,2,16)
print("*****************************")
#veo como esta mi inventario
print(mi_tienda.obtener_inventario())
print("*****************************")

#veo el diccionario de clientes
print(mi_tienda.obtener_clientes(con_compras=False))
print("*****************************")
#Pido el producto "Queso" por su nombre
print(mi_tienda.buscar_producto_por_nombre("Queso"))
print("*****************************")
#Calcular las ganancia
print(f"Tus ganacias hasta ahora: $", mi_tienda.calcular_ganancias())
print("*****************************")
#obtengo los clientes con mas compras
print("Tus mejores 5 clientes son:",mi_tienda.obtener_clientes_con_mas_compras(5))
print("*****************************")
```
